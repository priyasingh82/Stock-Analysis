# Beta calculation for past three months

```python

b= nf['Close'].pct_change()
nf['nf_chg']= b*100
nf['nf_chg'].iloc[0]=0
data2=nf.iloc[400:,7:8]
data2.head
```

```python
import statsmodels.api as sm
y=data2
x=data1
model=sm.OLS(y,x)
r=model.fit()
print(r.summary())
```

```python
# for every month
import statsmodels.api as sm
y=data2
x=data1
x.rolling(window=30)
y.rolling(window=30)
model=sm.OLS(y,x)
r=model.fit()
print(r.summary())
```
